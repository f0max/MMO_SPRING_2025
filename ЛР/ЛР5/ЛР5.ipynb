```python
text = 'В 2024 году в Москве прошла конференция по искусственному интеллекту, организованная компанией Яндекс.'
```

### 1. Токенизация

```python
!python -m spacy download ru_core_news_sm
```

```python
from spacy.lang.ru import Russian
import spacy


nlp = spacy.load('ru_core_news_sm')
spacy_text = nlp(text)
spacy_text
```

```python
for t in spacy_text:
    print(t)
```

### 2. Частеречная разметка

```python
for token in spacy_text:
    print(f'{token.text} - {token.pos_} - {token.dep_}')
```

### 3. Лемматизация

```python
for token in spacy_text:
      print(token, token.lemma, token.lemma_)
```

### 4. Выделение (распознавание) именованных сущностей.

```python
for ent in spacy_text.ents:
    print(ent.text, ent.label_)
```

```python
from spacy import displacy
displacy.render(spacy_text, style='ent', jupyter=True)
```

```python
print(spacy.explain("LOC"))
```

```python
print(spacy.explain("ORG"))
```
